```python
from diffusers import StableDiffusionPipeline
import torch

# Load the pre-trained model
model_id = "CompVis/stable-diffusion-v1-4"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

```python
import os

# Ensure the directory exists
os.makedirs('/content/drive/My Drive/Amgen', exist_ok=True)

# Set the file path to the desired directory
image_path = '/content/drive/My Drive/Amgen/generated_image.png'
```

```python
from PIL import Image
from IPython.display import display

# Generate an Image
prompt = "A red fedora hat floating in air magically"
image = pipe(prompt).images[0]

# Save the Image to the specified directory
image.save(image_path)

# Display the Image
image.show()

# Alternatively, display the image within the notebook
display(Image.open(image_path))
```
